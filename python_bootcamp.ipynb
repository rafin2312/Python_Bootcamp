Question 4(D)

```python
username = "admin"
password = "1234"
attempts = 0

while attempts <3:
  user_input_username = input("Enter your username: ")
  user_input_password = input("Enter your password: ")

  if user_input_username == username and user_input_password == password:
    print(f'Login successful Welcome, {user_input_username}.')
    break
  elif user_input_username == username and user_input_password != password:
    print('incorrect password, Try again')
  else:
    print('user not found.')
    attempts = attempts + 1
    if attempts == 3:
      print('Account locked.')
```

Question 5

```python
is_active = True
bank_user = "admin"
bank_password = "1234"

if is_active == False: # Corrected 'false' to 'False'
  print('Account is disable')
else: #if the account is active
  login_user = input('Enter your username: ')
  login_password = input('Enter your password: ')

  if login_user == bank_user and login_password == bank_password:
    print("Login successful")

    correct_pin = 1234
    balance = 5000
    pin_attempts = 3

    while pin_attempts > 0:
      entered_pin = int(input("Enter your pin: "))
      if entered_pin == correct_pin:
        request = int(input("Enter amount to withdraw: "))

        if request > 500 and request <= balance:
            balance = balance - request
            print(f"Successful withdrawal. Remaining balance: {balance}")
            break
        elif request <= 500:
            print("Amount must be greater than 500.")
        else: # request > balance
            print("Insufficient balance.")
        break
      else:
        pin_attempts = pin_attempts - 1
        if pin_attempts == 0:
            print('Card blocked!')
        else:
            print(f'Wrong pin. Attempts remaining: {pin_attempts}')
  else:
    print("Login failed.")
```
